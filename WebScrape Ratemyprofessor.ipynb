```python
from selenium import webdriver
from selenium.webdriver.support.select import Select
from bs4 import BeautifulSoup
import pandas as pd
import re
import numpy as np 
import json
import time 
import urllib
```

```python
alltid=pd.read_csv("./tid_list_full.csv")
```

```python
tid=alltid.iloc[:,0].tolist()
```

```python
prof_url= "http://www.ratemyprofessors.com/paginate/professors/ratings?tid={prof_id}&filter=&courseCode=&page={page_no}"
```

```python
import requests
baseurl = "http://www.ratemyprofessors.com/ShowRatings.jsp?tid={t_id}&showMyProfs=true"   # Change teach ID here

all_page_no=[]


```

```python
for i in range(0,len(tid)):
    print("getting information for",tid[i])
    teacher_id=tid[i]
    url=baseurl.format(t_id=teacher_id) # Change teach ID here
    print(url)
    page=requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser') #make the all reviews available for us
    try:
        numText = soup.find(class_="table-toggle rating-count active").get_text().strip() #get the number of reviews at the top of the website
        numOnly = int(re.findall('(\d+?)\s',numText)[0])  #get the integers of numText. \d means one digit. + means one or more. ? means 0 or 1, \s means white space.
        print("the num of reviews for",tid[i]," is ",numOnly)
    except:
        numOnly=0
        print("the num of reviews for",tid[i]," is ",numOnly)
    if numOnly!=0:
        all_page_no.append(numOnly//20+1)
    else:
        all_page_no.append(numOnly)
    time.sleep(1)
```

```python
print(len(all_page_no),len(tid))
```

```python
teacher_id=[]
attendance=[]
clarityColor=[]
easyColor=[]
helpColor=[]
helpCount=[]
userId=[]
nothelpCount=[]
onlineClass=[]
quality=[]
clarityRate=[]
className=[]
comments=[]
reviewDate=[]
easyRate=[]
helpfulRate=[]
interest=[]
overallRate=[]
status=[]
textBookUse=[]
wouldTakeAgain=[]
sId=[]
takenForCredit=[]
teacher=[]
finalGrade=[]
teacherTags=[]
unUsefulGrouping=[]
usefulGrouping=[]
```

```python
for i in range(0,len(tid)):
    if all_page_no[i]!=0:    
        for j in range(1,all_page_no[i]+1):
            professor_id=tid[i]
            current_page_no=j
            start_url=prof_url.format(prof_id=professor_id,page_no=current_page_no)
            print("getting ",j," page for professor ", professor_id)
            hdr = { 'User-Agent' : 'Mozilla/5.0 (Windows NT 6.1; Win64; x64)' }
            req = urllib.request.Request(start_url, headers=hdr)
            response = urllib.request.urlopen(req).read().decode()
            data = json.loads(response)
            time.sleep(5)
            ratings=data["ratings"]
            for k in range(0,len(ratings)):
                teacher_id.append(tid[i])
                attendance.append(ratings[k]["attendance"])
                clarityColor.append(ratings[k]["clarityColor"])
                easyColor.append(ratings[k]['easyColor'])
                helpColor.append(ratings[k]['helpColor'])
                helpCount.append(ratings[k]['helpCount'])
                userId.append(ratings[k]['id'])
                nothelpCount.append(ratings[k]['notHelpCount'])
                onlineClass.append(ratings[k]['onlineClass'])
                quality.append(ratings[k]['quality'])
                clarityRate.append(ratings[k]['rClarity'])
                className.append(ratings[k]['rClass'])
                comments.append(ratings[k]['rComments'])
                reviewDate.append(ratings[k]['rDate'])
                easyRate.append(ratings[k]['rEasy'])
                helpfulRate.append(ratings[k]['rHelpful'])
                interest.append(ratings[k]["rInterest"])
                overallRate.append(ratings[k]["rOverall"])
                status.append(ratings[k]["rStatus"])
                textBookUse.append(ratings[k]["rTextBookUse"])
                wouldTakeAgain.append(ratings[k]["rWouldTakeAgain"])
                sId.append(ratings[k]["sId"])
                takenForCredit.append(ratings[k]["takenForCredit"])
                teacher.append(ratings[k]["teacher"])
                finalGrade.append(ratings[k]["teacherGrade"])
                teacherTags.append(ratings[k]["teacherRatingTags"])
                unUsefulGrouping.append(ratings[k]["unUsefulGrouping"])
                usefulGrouping.append(ratings[k]["usefulGrouping"])
```
